```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("HR-Employee-Attrition.csv")

# Basic information
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

# Attrition Summary
attrition_count = df['Attrition'].value_counts()
print("\nAttrition Count:")
print(attrition_count)

# Attrition Percentage
attrition_percentage = (attrition_count / len(df)) * 100
print("\nAttrition Percentage:")
print(attrition_percentage)

# Attrition by Department
dept_attrition = pd.crosstab(
    df['Department'],
    df['Attrition'],
    normalize='index'
) * 100

print("\nAttrition by Department (%):")
print(dept_attrition)

# Average Monthly Income by Attrition Status
income_analysis = df.groupby('Attrition')['MonthlyIncome'].mean()
print("\nAverage Monthly Income:")
print(income_analysis)

# Visualization
attrition_count.plot(
    kind='bar',
    title='Employee Attrition',
    ylabel='Number of Employees'
)

plt.tight_layout()
plt.show()
```
